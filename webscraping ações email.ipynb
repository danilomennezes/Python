```python
#Buscando Dados fundamentalista
import pandas as pd
cam_arquivo ='C:/Users/danil/Downloads/statusinvest-busca-avancada.csv'
df = pd.read_csv(cam_arquivo, sep=';')
df
```

```python
#Separando os dados fundamentalista das ações
FBBDC3 = df.loc[df['TICKER']=='BBDC3']
PLBBDC3   = FBBDC3['P/L'].values
DYBBDC3   = FBBDC3['DY'].values
P_VPBBDC3 = FBBDC3['P/VP'].values
EV_BITBBDC3 = FBBDC3['EV/EBIT'].values
```

```python
aPLBBDC3 = f'O P/L (Preço/Lucro) do Bradesco é de {PLBBDC3[0]}. Isso indica que o preço atual da ação está {PLBBDC3[0]} vezes o lucro por ação.'
aDYBBDC3 = f'O dividend yield do Bradesco é de {DYBBDC3[0]}%. Isso indica qual percentual do valor atual da ação é distribuído aos acionistas na forma de dividendos.'
aP_VPBBDC3 = f'O preço sobre o valor patrimonial (P/VP) do Bradesco é de {P_VPBBDC3[0]}. Isso sugere que o preço atual da ação está sendo negociado a {P_VPBBDC3[0]} do seu valor patrimonial.' 
aEV_BITBBDC3 = f'O EV/EBIT (Enterprise Value / Earnings Before Interest and Taxes) do Bradesco é de {EV_BITBBDC3[0]}. Isso compara o valor da empresa (considerando dívidas e caixa) com o seu lucro antes de juros e impostos, indicando a eficiência da empresa em gerar lucro em relação ao seu valor total.'
```

```python
#Yfinance - Buscando cotação das ações 
import yfinance as yf
bradesco = yf.download("BBDC3.SA", start = "2024-01-01")
#pegando os valores de fechamento
a = bradesco['Adj Close'].values
b = len(a) -1
c = len(a) -2
```

```python
#Verificando se houve aumento ou decrécimo
if (((a[b]-a[c])/a[c])*100) < 0:
  var1 = f'Hoje, a ação do Bradesco teve uma diminuição de {(((a[b]-a[c])/a[c])*100):.2f}%. Isso indica que o preço da ação caiu.'
else:
  var1 = f'Hoje, a ação do Bradesco teve uma aumento de {(((a[b]-a[c])/a[c])*100):.2f}%. Isso indica que o preço da ação subiu.'


if (((a[b]-a[0])/a[0])*100) < 0:
  var2 = f'No acumulado do ano, a ação do Bradesco teve um decréscimo de {(((a[b]-a[0])/a)[0]*100):.2f}%. Isso sugere que o desempenho da ação ao longo do ano foi negativo.'
else:
  var2 = f'No acumulado do ano, a ação do Bradesco teve um aumento de {(((a[b]-a[0])/a)[0]*100):.2f}%. Isso sugere que o desempenho da ação ao longo do ano foi negativo.'

print(f'{var1}, {var2} ')
```

```python
import smtplib
import email.message
from email.mime.base import MIMEBase
from email import encoders
```

```python
def enviar_email():
    body = f'<h1>Bradesco BBDC3</h1> <h3>Variação de Preço da Ação:</h3> <p>{var1}</p> <h3>Desempenho Anual:</h3> <p>{var2}</p> <h3>P/L (Preço/Lucro):</h3> <p>{aPLBBDC3}</p> <h3>Preço sobre o Valor Patrimonial:</h3> <p>{aP_VPBBDC3}</p><h3>Dividend Yield:</h3> <p>{aDYBBDC3}</p><h3>EV/EBIT:</h3> <p>{aEV_BITBBDC3}</p>'
                                            
    
    msg = email.message.Message()
    msg['Subject'] = "Ações Analise Bradesco (BBDC3)"
    msg['From'] = 'danilomennezes@gmail.com'
    msg['To'] = 'danilomennezes@gmail.com'
    password = 'colocar aqui sua senha'
    msg.add_header('Content-Type', 'text/html')
    msg.set_payload(body)
   
    
    s = smtplib.SMTP('smtp.gmail.com: 587')
    s.starttls()
    s.login(msg['From'], password)
    s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
    print('Email enviado')
```
